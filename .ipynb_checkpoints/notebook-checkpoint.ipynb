## Imports: <a class = "anchor" id = "imports"></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

## Data: <a class = "anchor" id = "data"></a>

```python
df = pd.read_csv("stats.csv",).rename(columns = {" first_name" : "first_name"})
df["first_name"] = df["first_name"].str.strip()
df["full_name"] = df["first_name"] + " " + df["last_name"]
df.insert(0, "full_name", df.pop("full_name"))
df.drop(columns = "Unnamed: 38", inplace = True)
df["slash_line"] = df["batting_avg"].astype(str).map(lambda x: x[1:5]) + "/" +\
                   df["on_base_percent"].astype(str).map(lambda x: x[1:5]) + "/" +\
                   df["slg_percent"].astype(str).map(lambda x: x[1:5]) + "/" +\
                   df["on_base_plus_slg"].astype(str).map(lambda x: x[1:5])
```

## EDA: <a class = "anchor" id = "eda"></a>

```python
# Relationship between exit velocity and launch angle
x = df["exit_velocity_avg"]
y = df["launch_angle_avg"]


plt.figure(figsize = (12,8))
plt.scatter(x,y, c = "b")

z = np.polyfit(x,y,1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")

plt.title("2021 Avg. Exit Velocity vs Avg. Launch Angle (as of 7/29/21)")
plt.xlabel("Average Exit Velocity")
plt.ylabel("Average Launch Angle")

a = df.loc[df["exit_velocity_avg"] == min(df["exit_velocity_avg"]), ["full_name", "exit_velocity_avg", "launch_angle_avg"]].reset_index()
plt.annotate(a["full_name"][0],
             (a["exit_velocity_avg"][0], a["launch_angle_avg"][0]),
             ha = "center",
             va = "bottom")

b = df.loc[df["exit_velocity_avg"] == max(df["exit_velocity_avg"]), ["full_name", "exit_velocity_avg", "launch_angle_avg"]].reset_index()
plt.annotate(b["full_name"][0],
             (b["exit_velocity_avg"][0], b["launch_angle_avg"][0]),
             ha = "center",
             va = "bottom")

c = df.loc[df["launch_angle_avg"] == min(df["launch_angle_avg"]), ["full_name", "exit_velocity_avg", "launch_angle_avg"]].reset_index()
plt.annotate(c["full_name"][0],
             (c["exit_velocity_avg"][0], c["launch_angle_avg"][0]),
             ha = "center",
             va = "bottom")

d = df.loc[df["launch_angle_avg"] == max(df["launch_angle_avg"]), ["full_name", "exit_velocity_avg", "launch_angle_avg"]].reset_index()
plt.annotate(d["full_name"][0],
             (d["exit_velocity_avg"][0], d["launch_angle_avg"][0]),
             ha = "center",
             va = "bottom");
# plt.savefig("plot.png");
```
